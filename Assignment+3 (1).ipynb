```python
#Explore the data
import pandas as pd
import numpy as np
from pandas.tools.plotting import scatter_matrix
%matplotlib inline
#Read iris.txt into a data frame (2 point)
FIN = pd.read_csv('iris.txt')
#Print the head the dataset (2 point)
print(FIN.head(1))
print("-"*125)
#Print the tail of the dataset (2 point)
print(FIN.tail(1))
```

Between the first observation and final observation I don't see significant changes in the sepal length or width. However, there does appear to be noticable change in the petal length and width. However, this information is not enough to use to draw conclusions.

```python
#Print the summary statistics for the frame (2 point)
import pandas as pd
import numpy as np
from pandas.tools.plotting import scatter_matrix
%matplotlib inline
FIN = pd.read_csv('iris.txt')
print(FIN.describe())
print(FIN.describe(1))
```

```python
#Generate a box plot for the data set (2 point)
import pandas as pd
import numpy as np
from pandas.tools.plotting import scatter_matrix
%matplotlib inline
FIN = pd.read_csv('iris.txt')
print(FIN.boxplot(return_type='axes'))
```

In the boxplot I can see that petal length has a wide range of values across observations and that these values extend far below the mean.

```python
#Generate a line plot for dataset
import pandas as pd
import numpy as np
from pandas.tools.plotting import scatter_matrix
%matplotlib inline
FIN = pd.read_csv('iris.txt')
print(FIN.plot())
```

From this graph, I see interesting "jumps" in the petal length that occur between groups of observations. It also appears that petal width also has "jumps" as well, though not as significant as petal length. I do not see this behavior in sepal length or width.

```python
#Generate a histogram of the dataset (2 point)
import pandas as pd
import numpy as np
from pandas.tools.plotting import scatter_matrix
%matplotlib inline
FIN = pd.read_csv('iris.txt')
print(FIN.hist(xlabelsize=8,ylabelsize=8,bins=9))
```

I can see in the petal length and width graphs that there are significant outliers that cause the curve to be abnormal. Sepal length and sepal width follow relatively normal curves.

```python
#I want to see if the different flower species have significant enought differences to enable me to use them for classification
import pandas as pd
import numpy as np
from pandas.tools.plotting import scatter_matrix
%matplotlib inline
#I created a new set that was grouped by species to get more meaningful summary statistics
NFIN=FIN.groupby("Species")
print(NFIN.describe())
```

```python
#Generate a box plot for the data set (2 point)
import pandas as pd
import numpy as np
from pandas.tools.plotting import scatter_matrix
%matplotlib inline
NFIN=FIN.groupby("Species")
print(NFIN.boxplot(return_type='axes',layout=(3,1),figsize=(6,7)))
```

Based on this box plot you can clearly see that the most distinguishable factors among these species is petal length and width. In fact, none of the observations have overlapping ranges in these measurements.

```python
#Generate a histogram of the dataset (2 point)
import pandas as pd
import numpy as np
from pandas.tools.plotting import scatter_matrix
%matplotlib inline
NFIN=FIN.groupby("Species")
print(NFIN.hist(bins=9))
```

When the data is broken down by species, the curves no longer appear as "normal" as they seemed before. There are multiple modes or clusters within each of the species that are affecting the mean.
